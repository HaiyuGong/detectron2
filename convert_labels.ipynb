```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image, ImageDraw
import json
```

```python
img_path=r'/hy-tmp/SHIDC-B-Ki-67/256x256-cropped-images/train256/294_0031_5.jpg'
img = Image.open(img_path)
with open(img_path.replace('.jpg', '.json')) as f:
    points=json.load(f)
draw = ImageDraw.Draw(img)
radius = 4
for cell in points:
    if cell['label_id']==1:
        # draw.ellipse((cell['x'] - radius, cell['y'] - radius, cell['x'] + radius, cell['y'] + radius), outline="red")
        draw.rectangle([cell['x'] - radius+1, cell['y'] - radius+1, cell['x'] + radius+1, cell['y'] + radius+1], outline="red")
    elif cell['label_id']==2:
        # draw.ellipse((cell['x'] - radius, cell['y'] - radius, cell['x'] + radius, cell['y'] + radius), outline="blue")
        draw.rectangle([cell['x'] - radius, cell['y'] - radius, cell['x'] + radius, cell['y'] + radius], outline="blue")
    elif cell['label_id']==3:
        # draw.ellipse((cell['x'] - radius, cell['y'] - radius, cell['x'] + radius, cell['y'] + radius), outline="green")
        draw.rectangle([cell['x'] - radius, cell['y'] - radius, cell['x'] + radius, cell['y'] + radius], outline="green")
    
img = np.array(img)
print(img.shape)
plt.imshow(img)
plt.show()
```
